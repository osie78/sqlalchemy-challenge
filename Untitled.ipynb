```python
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
```

```python
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine
from sqlalchemy import func
import pandas as pd
import datetime as dt
from datetime import date
import numpy as np
```

```python
# Create engine using the `demographics.sqlite` database file
engine = create_engine("sqlite:///hawaii.sqlite")
```

```python
# Declare a Base using `automap_base()`
Base = automap_base()
```

```python
# Use the Base class to reflect the database tables
Base.prepare(engine, reflect=True)
```

```python
# View all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement=Base.classes.measurement
Station=Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session=Session(engine)
```

# Exploratory Precipitation Analysis

```python
first_row = session.query(Measurement).first()
first_row.__dict__
```

```python
q_measurement=session.query(Measurement)
```

```python
last_measurement=session.query(Measurement.date).order_by(Measurement.date.desc()).first()
print(last_measurement)
```

```python
print(f'the most recent date is {last_measurement}')
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results. 
date1=[]
prec1=[]
query_12_months = q_measurement.filter(Measurement.date.between('2016-08-23', '2017-08-23'))
for x in query_12_months:
    date1.append(x.date)
    prec1.append(x.prcp)
    print(x.date,x.prcp)
```

```python
date1_df=pd.DataFrame(date1)
prec1_df=pd.DataFrame(prec1)
```

```python
date1_df['Date']=pd.DataFrame(date1)
prec1_df['Precipitation']=pd.DataFrame(prec1)

prec_1y_df= pd.concat([date1_df,prec1_df], axis=1)
del prec_1y_df[0]

prec_1y_df=prec_1y_df.dropna(subset=['Precipitation'])
prec_1y_df=prec_1y_df.sort_values(by=['Date'])
prec_1y_df
```

```python
prec_1y_df.plot(x='Date',y='Precipitation')
plt.xticks(rotation='vertical')
plt.legend(loc=(.3,1))
plt.show()
```

#### Statistical analysis

```python
avg_p=prec_1y_df['Precipitation'].mean()
med_p=prec_1y_df['Precipitation'].median()
st_p=prec_1y_df['Precipitation'].std()
print(f'the mean precipitation in the 12 months period was {avg_p}')
print(f'the median precipitation in the 12 months period was {med_p}')
print(f'the standard deviation in the 12 months period was {st_p}')
```

## Station Analysis

```python
# Design a query to calculate the total number stations in the dataset
f_row = session.query(Station).first()
f_row.__dict__
```

```python
number_stations=session.query(Station.name).count()
print(f'there are {number_stations} stations reporting')
```

```python
count_per_station=session.query(Measurement.station, func.count(Measurement.station)).group_by(Measurement.station).all()
count_per_station_df=pd.DataFrame.from_records(count_per_station, columns=['station', 'records'])

count_per_station_df=count_per_station_df.sort_values(by=['records'], ascending=False)
count_per_station_df
```

```python
most_act=count_per_station_df['station'].iloc[0]
print(f'the station with the highest activity is {most_act}')
```

### Calculating min, max and average temperature for the most active station

```python
q_temp_max=session.query(Measurement.tobs,func.max(Measurement.tobs)).filter_by(station=most_act).all()

print(f' the max observed temp is {q_temp_max[0][0]}')
```

```python
q_temp_min=session.query(Measurement.tobs,func.min(Measurement.tobs)).filter_by(station=most_act).all()
print(f' the min observed temp is {q_temp_min[0][0]}')
```

```python
q_temp_avg=session.query(Measurement.tobs,func.avg(Measurement.tobs)).filter_by(station=most_act).all()
print(f' the average observed temp is {q_temp_avg[0][0]}')
```

### Design a query to retrieve the last 12 months of temperature observation data (TOBS).

```python
date2=[]
temp1=[]
station=[]
query_12_months = q_measurement.filter(Measurement.date.between('2016-08-23', '2017-08-23'))

for y in query_12_months:
    date2.append(y.date)
    station.append(y.station)
    temp1.append(y.tobs)
    print(y.date,y.station,y.tobs)
```

```python
date2_df=pd.DataFrame(date2)
date2_df['Date']=pd.DataFrame(date2_df)

temp1_df=pd.DataFrame(temp1)
temp1_df['Temp']=pd.DataFrame(temp1_df)

station_df=pd.DataFrame(station)
station_df['Station']=pd.DataFrame(station_df)

temp_1y_df= pd.concat([date2_df,temp1_df,station_df], axis=1)
del temp_1y_df[0]

temp_1y_df=temp_1y_df.dropna(subset=['Temp']) # deleting null values in temperatures

temp_1y_df
```

```python
most_act_temp=temp_1y_df.groupby('Station').count()
most_act_temp[['Temp']]
```

```python
t_most_act_station=temp_1y_df.loc[(temp_1y_df['Station'] == 'USC00513117')].Temp
t_most_act_station_df=pd.DataFrame(t_most_act_station)
t_most_act_station_df
```

```python

hist=t_most_act_station_df.hist(bins=12)

plt.ylabel('Frequency')
plt.xlabel('Temperature')
plt.show()
```
